```lua
torch.setdefaulttensortype('torch.FloatTensor')
function rotatePoint(radian, point, origin)
    local rx = math.cos(radian) * (point[1] - origin[1]) - math.sin(radian) * (point[2] - origin[2]) + origin[1]
    local ry = math.sin(radian) * (point[1] - origin[1]) + math.cos(radian) * (point[2] - origin[2]) + origin[2]
    return rx, ry
end
function markImage(img, point)
    for i = 1, 3 do
        img[i][point[1]][point[2]] = 1
    end
    return img
end
```

```lua
w, h = 200, 200
img = image.load('boer.jpeg')
img = image.scale(img, w, h)

imgdot = img:clone()
imgdot = image.rotate(imgdot, math.rad(270), 'bilinear')
point = {25, 150}
markImage(imgdot, point)
imgdot = image.rotate(imgdot, math.rad(-270), 'bilinear')
```

```lua
itorch.image(img), itorch.image(imgdot)
```

```lua
radian = math.rad(45)
point1 = {20, 20}
point2 = {40, 40}
origin = {30, 30}
rx1, ry1 = rotatePoint(radian, point1, origin)
rx2, ry2 = rotatePoint(radian, point2, origin)
print(rx1, ry1, rx2, ry2)
```
