```lua
train = torch.load('train.t7');
valid = torch.load('valid.t7');
```

```lua
index = {};
for i = 1, #train do
    table.insert(index, i)
end
```

```lua
plot = itorch.Plot()
plot:line(index, train, 'red', 'train')
plot:line(index, valid, 'blue', 'valid')
plot:title('Accuracy over training')
plot:xaxis('epoch'):yaxis('accuracy')
plot:legend(true)
plot:draw()
```
